```python
#import dependencies
import pandas as pd
```

```python
#import the data and drop unwanted rows and columns
pd.set_option('max_colwidth', 400)
adidas_sales_df = pd.read_excel("Adidas US Sales Datasets.xlsx", skiprows=4)[:-2]

# Drop the null columns where all values are null
adidas_sales_df = adidas_sales_df.dropna(axis='columns', how='all')

# Drop the null rows
adidas_sales_df = adidas_sales_df.dropna()
```

```python
#view the dataframe
adidas_sales_df.head()
```

```python
#drop incorrectly calculated total sales and operating profit columns
updated_sales_df = adidas_sales_df.drop("Total Sales", axis='columns')
updated_sales_df2 = updated_sales_df.drop("Operating Profit", axis='columns')
```

```python
updated_sales_df2.head()
```

```python
#drop city columns for region and state focus of data
adidas_sales_df = updated_sales_df2.drop("City", axis='columns')
adidas_sales_df.head()
```

```python
#Regerating new Total Sales column
adidas_sales_df['Total Sales'] = adidas_sales_df['Price per Unit'] * adidas_sales_df['Units Sold']


#Regerating new Operating Profit column
adidas_sales_df['Operating Profit'] = adidas_sales_df['Total Sales'] * adidas_sales_df['Operating Margin']

adidas_sales_df.head()
```

```python
adidas_sales_df.describe()
```

```python
# Export the DataFrame as a CSV file using encoding='utf8'.
output_data_file = "cleaned_adidas_sales_data.csv"
adidas_sales_df.to_csv(output_data_file, encoding='utf8', index= False)
```
