```python
import pandas as pd
import numpy as np
```

```python
#Import the CSV File

adidas_sales_df = pd.read_csv("cleaned_adidas_sales_data.csv")

#Read the file
adidas_sales_df
```

```python
#Retrieve Unique Products

products = (adidas_sales_df["Product"].unique())

#Create Unique IDS for Products

Product_ids = np.arange(1, 7)
p_ids = ["PD" + str(p_ids) for p_ids in Product_ids]


#New DataFrame

Products_df = pd.DataFrame({
    "Product_ID": p_ids,
    "Product" : products})


#Display the Data Frame
Products_df

```

```python
#Save the Date Frame to CSV File

Products_df.to_csv('Products.csv',index = False)
```

```python
#Retrieve Unique Regions
regions = (adidas_sales_df["Region"].unique())
regions_ids = np.arange(1, 6)
r_ids = ["RG" + str(r_ids) for r_ids in regions_ids]

#New DataFrame

Regions_df = pd.DataFrame({
    "Region_ID": r_ids,
    "Region" : regions})

#Set Index
#Regions_df = Regions_df.set_index(['Region_ID'])

#Display the Data Frame
Regions_df
```

```python
Regions_df.to_csv('Regions.csv',index = False)
```

```python
#Retreive Unique States

states = (adidas_sales_df["State"].unique())

#Create Unique IDS for States

states_ids = np.arange(1, 51)
s_ids = ["ST" + str(s_ids) for s_ids in states_ids]

#New DataFrame

States_df = pd.DataFrame({
    "State_ID": s_ids,
     "State":states
                      }
)


```

```python
#Reterive the Data Frame

States_df
```

```python
Regions_df
```

```python
# Merge Product Id, State ID, Region ID to main Orginial Data Frame i.e, adidas_sales_df


merged_df = adidas_sales_df.merge(Regions_df, on='Region', how='left').merge(Products_df, on = 'Product', how ='left').merge(States_df, on = 'State', how = 'left')
merged_df.head(100)
```

```python
#Select multiple Columns and saving into a Data Frame
States_df = merged_df[['State','State_ID','Region_ID']].drop_duplicates()

#Display Data Frame
States_df
```

```python
States_df.to_csv('States.csv',index = False)
```

```python
#Select Multiple Columns from merged_df into a new Data Frame

retailers = merged_df[['Retailer ID','Retailer']].drop_duplicates()

#View the Data Frame
retailers
```

```python
#Save Retailers Data frame to CSV

retailers.to_csv('Retailers.csv',index = False)
```

```python
#Select Multiple Columns from merged_df into a new Data Frame

adidas_sales = merged_df[['Retailer ID','Invoice Date','Region_ID','State_ID','Product_ID','Price per Unit','Units Sold',
                           'Operating Margin','Sales Method','Total Sales','Operating Profit']]
```

```python
#Display The Data Frame

adidas_sales
```

```python
#Saving the Date Frame to CSV File

adidas_sales.to_csv('Adidas_sales.csv',index = False)
```
