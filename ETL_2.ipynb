```python
import pandas as pd
import numpy as np
```

```python
#Import the CSV File

adidas_sales_df = pd.read_csv("cleaned_adidas_sales_data.csv")

#Read the file
adidas_sales_df
```

```python
#Drop the Retailer ID Column

adidas_sales_df.drop(['Retailer ID'], axis=1, inplace = True , errors = 'ignore')
adidas_sales_df
```

```python
#Retrieve Unique Products

products = (adidas_sales_df["Product"].unique())

#Create Unique IDS for Products

Product_ids = np.arange(1, 7)
p_ids = ["PD" + str(p_ids) for p_ids in Product_ids]


#New DataFrame

Products_df = pd.DataFrame({
    "Product_ID": p_ids,
    "Product" : products})


#Display the Data Frame
Products_df

```

```python
#Save the Date Frame to CSV File

Products_df.to_csv('Products.csv',index = False)
```

```python
#Retrieve Unique Regions
regions = (adidas_sales_df["Region"].unique())
regions_ids = np.arange(1, 6)
r_ids = ["RG" + str(r_ids) for r_ids in regions_ids]

#New DataFrame

Regions_df = pd.DataFrame({
    "Region_ID": r_ids,
    "Region" : regions})

#Set Index
#Regions_df = Regions_df.set_index(['Region_ID'])

#Display the Data Frame
Regions_df
```

```python
Regions_df.to_csv('Regions.csv',index = False)
```

```python
#Retreive Unique States

retailers = adidas_sales_df["Retailer"].unique()


#Create Unique IDS for States

retailer_id = np.arange(1, 7)
rtl_ids = ["RTL" + str(s_ids) for s_ids in retailer_id]

#New DataFrame

retailer_df = pd.DataFrame({
    "Retailer_ID": rtl_ids ,
'Retailer' : retailers})

retailer_df
```

```python
# Save Data Frame to CSV fILE
retailer_df.to_csv('Retailers.csv',index = False)
```

```python
#Retreive Unique States

states = (adidas_sales_df["State"].unique())

#Create Unique IDS for States

states_ids = np.arange(1, 51)
s_ids = ["ST" + str(s_ids) for s_ids in states_ids]

#New DataFrame

States_df = pd.DataFrame({
    "State ID": s_ids,
     "State":states
                      }
)
```

```python
#Reterive the Data Frame

States_df
```

```python
Regions_df
```

```python
# Merge Product Id, State ID, Region ID to main Orginial Data Frame i.e, adidas_sales_df


merged_df = (adidas_sales_df.merge(Regions_df, on='Region', how='left')
                           .merge(Products_df, on = 'Product', how ='left')
                           .merge(States_df, on = 'State', how = 'left')
                           .merge(retailer_df, on ='Retailer', how ='left'))

merged_df.head(100)
```

```python
#Select multiple Columns and saving into a Data Frame
States_df = merged_df[['State','State ID','Region_ID']].drop_duplicates()

#Display Data Frame
States_df
```

```python
#Import the CSV File

us_population_df = pd.read_excel('/Users/ranjodh/Desktop/Data Analytics Bootcamp/Final-Project/US_Population_Data/US_Population_Data_Summary.xlsx')

#Read the file
us_population_df
```

```python
# Merge State DF With US Population DF

States_df = us_population_df.merge(States_df , on = ['State ID', 'State'], how = 'left')


States_df.drop(['State_y','Male population','Female Population','Male median age','Female median age',
                           'Population 16 years and over with earnings'], axis=1, 
         inplace = True, errors ='ignore')

# #Display Data Frame
States_df


```

```python
#Rename the States Data Frame

States_df.rename(columns={'State ID' :'State_ID','Total population':'Population','Median age (years)': 
                                      'Median_age','Sex ratio (males per 100 females)' : 'Sex_ratio','Median earnings (dollars)'
                                      : 'Median_income'}, inplace = True)


#Reindexing the States_df
States_df =States_df.reindex(['Region_ID','State_ID','State','Population','Median_age','Sex_ratio','Median_income']
                             , axis=1)

States_df


```

```python
#CSV File
States_df.to_csv('States.csv',index = False)
```
