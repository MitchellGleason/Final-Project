```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("AdidasSales").config("spark.driver.extraClassPath","/content/postgresql-42.5.1.jar").getOrCreate()
```

```python
# Read in data from S3 Buckets
from pyspark import SparkFiles
url="https://adidasdatag13.s3.us-east-2.amazonaws.com/Adidas_sales.csv"
spark.sparkContext.addFile(url)
Adidas_sales_df = spark.read.csv(SparkFiles.get("Adidas_sales.csv"), sep=",", header=True, inferSchema=True)

# Show DataFrame
Adidas_sales_df.show()
```

```python
url="https://adidasdatag13.s3.us-east-2.amazonaws.com/Tableau.csv"
spark.sparkContext.addFile(url)
Tableau_df = spark.read.csv(SparkFiles.get("Tableau.csv"), sep=",", header=True, inferSchema=True)

# Show DataFrame
Tableau_df.show()
```
